1.머신러닝  
-> 라벨별 데이터의 비율이 차이가 많이 날 경우   
    ex) 0 : 99%, 1 : 0.01% -> 1라벨이 거의 포함되지 않을 수 있음.
        -> Oversampling : Smote : 1라벨에 7%만 오버 샘플링
        -> 실제로는 사용 X, 변수가 많으면 엉뚱한 데이터가 생성됌.  
  
2.CNN  
-> 이미지데이터 : (이미지는 약간 변형 가능)

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Input, Dense, Dropout
```

1층  
뉴런의 수 : 100개  
활성화 함수 : sigmoid  
드롭아웃 : 20%  

2층  
뉴런의 수 : 200개  
활성화 함수 : relu  
드롭아웃 : 40%  

3층  
뉴런의 수 : 50개  
활성화 함수 : tanh  
드롭아웃 : 10%  

아웃풋  
뉴런의 수 : 6개  
활성화 함수 : softmax  


결과값.  
손실 함수 : 크로스엔트로피  
옵티마이저 : adam, 학습률 : 0.001  
배치 크기 : 200  
학습 횟수 : 300  

```python
model = Sequential([
    Input(shape=(11,)),
    Dense(100,activation='sigmoid'),
    Dropout(0.2),
    Dense(200,activation='relu'),
    Dropout(0.4),
    Dense(50,activation='tanh'),
    Dropout(0.1),
    Dense(10,activation='softmax') #결과가 3~8등급임 -> 0~5등급 즉 6으로 설정해도 괜찮음. -> 그렇다면 와이값들을 -3해줘서 넣어줘야함.
])
model.summary()
```

```python
import pandas as pd
redwine = pd.read_csv('winequality-red.csv', sep = ';')
X,y = redwine.iloc[:,:-1],redwine.iloc[:,-1]
```

```python
redwine
```

```python
#학습데이터와 테스트데이터를 나눔.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
#훈련정의
model.compile(loss='sparse_categorical_crossentropy',optimizer='adam', metrics=['accuracy'])
#cross -> classification
#sparse : 원핫인코딩 활용
```

```python
#데이터 학습
history = model.fit(X_train, y_train, batch_size=200, epochs=300)
```

```python
model.evaluate(X_test,y_test)
```
